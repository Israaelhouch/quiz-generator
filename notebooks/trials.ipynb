```python
from pathlib import Path
from src.embeddings.utils import load_runtime_config
from src.embeddings.build_vector_store import build_vector_store
from src.retrieval import Retriever

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

# 1) load config
config = load_runtime_config(ROOT / "configs/config.yaml")
```

```python
# 2) build all profiles (auto)
for profile_name in config.profiles.keys():
    cfg = config.with_profile(profile_name)
    info = build_vector_store(cfg)
    print(profile_name, info["rows_indexed"], info["collection_name"])
```

```python
# 3) retriever

# Arabic -> arabic profile 
r = Retriever(config_path=config)
ar_results= r.retrieve(
    query="التّشبِيه المَقْلُوبُ",
    language="ar",
    subjects=["ARABIC"],
    top_k=10,
)

for q in ar_results:
    print("distance:", q.distance)
    print("search_text:", q.search_text)
    print("------")
```
